```python
# CryptoAdvisor Chatbot
def crypto_advisor():
    # Predefined crypto dataset
    crypto_db = {
        "Bitcoin": {
            "price_trend": "rising",
            "market_cap": "high",
            "energy_use": "high",
            "sustainability_score": 3/10
        },
        "Ethereum": {
            "price_trend": "stable",
            "market_cap": "high",
            "energy_use": "medium",
            "sustainability_score": 6/10
        },
        "Cardano": {
            "price_trend": "rising",
            "market_cap": "medium",
            "energy_use": "low",
            "sustainability_score": 8/10
        },
        "Solana": {
            "price_trend": "rising",
            "market_cap": "medium",
            "energy_use": "low",
            "sustainability_score": 7/10
        },
        "Polygon": {
            "price_trend": "stable",
            "market_cap": "medium",
            "energy_use": "low",
            "sustainability_score": 7/10
        }
    }

    # Chatbot introduction
    print("\nWelcome to CryptoBuddy! ðŸ¤–ðŸ’°")
    print("Your friendly AI-powered cryptocurrency advisor.")
    print("I can help you find cryptocurrencies based on:")
    print("- Profitability (price trends, market cap)")
    print("- Sustainability (energy efficiency, project viability)")
    print("Type 'quit' to exit at any time.\n")
    print("Some questions you can ask me:")
    print("- Which crypto is trending up?")
    print("- What's the most sustainable coin?")
    print("- What should I buy for long-term growth?")
    print("- Show me eco-friendly cryptos")
    print("- Recommend a high market cap coin\n")

    # Main chat loop
    while True:
        user_query = input("\nYou: ").lower()

        if user_query == 'quit':
            print("CryptoBuddy: Happy investing! Remember to do your own research too. ðŸš€")
            break

        # Response logic
        elif any(word in user_query for word in ['trend', 'rising', 'growing', 'increasing']):
            trending_coins = [coin for coin in crypto_db if crypto_db[coin]["price_trend"] == "rising"]
            if trending_coins:
                print(f"CryptoBuddy: These cryptos are currently trending up: {', '.join(trending_coins)}")
                best_trend = max(trending_coins, key=lambda x: crypto_db[x]["sustainability_score"])
                print(f"  My top pick from these is {best_trend} with sustainability score {crypto_db[best_trend]['sustainability_score']*10}/10 ðŸŒ±")
            else:
                print("CryptoBuddy: No cryptocurrencies are currently showing strong upward trends.")

        elif any(word in user_query for word in ['sustainab', 'eco', 'green', 'environment']):
            sustainable_coins = sorted(
                crypto_db.keys(),
                key=lambda x: crypto_db[x]["sustainability_score"],
                reverse=True
            )
            print("CryptoBuddy: Here are cryptos ranked by sustainability:")
            for coin in sustainable_coins[:3]:  # Show top 3
                score = crypto_db[coin]["sustainability_score"] * 10
                print(f"  {coin}: {score}/10 (Energy use: {crypto_db[coin]['energy_use']})")

        elif any(word in user_query for word in ['long-term', 'long term', 'future', 'growth']):
            # Balance between sustainability and positive trend
            best_long_term = max(
                crypto_db.keys(),
                key=lambda x: (crypto_db[x]["sustainability_score"] * 0.7 +
                              (0.3 if crypto_db[x]["price_trend"] == "rising" else 0))
            )
            print(f"CryptoBuddy: For long-term growth, consider {best_long_term}! ðŸ’¹")
            print(f"  It has a sustainability score of {crypto_db[best_long_term]['sustainability_score']*10}/10")
            print(f"  and is currently {'rising' if crypto_db[best_long_term]['price_trend'] == 'rising' else 'stable'} in price.")

        elif any(word in user_query for word in ['market cap', 'large cap']):
            high_cap_coins = [coin for coin in crypto_db if crypto_db[coin]["market_cap"] == "high"]
            if high_cap_coins:
                print(f"CryptoBuddy: High market cap cryptos: {', '.join(high_cap_coins)}")
                print("  These are generally more stable but may have slower growth.")
            else:
                print("CryptoBuddy: No high market cap cryptos in my database currently.")

        elif any(word in user_query for word in ['recommend', 'suggest', 'advice', 'what should']):
            # Balanced recommendation considering both factors
            recommended = max(
                crypto_db.keys(),
                key=lambda x: (crypto_db[x]["sustainability_score"] * 0.6 +
                              (0.4 if crypto_db[x]["price_trend"] == "rising" else 0) +
                              (0.2 if crypto_db[x]["market_cap"] == "high" else 0.1))
            )
            details = crypto_db[recommended]
            print(f"CryptoBuddy: I recommend {recommended}! ðŸ“ˆ")
            print(f"  Price trend: {details['price_trend'].capitalize()}")
            print(f"  Market cap: {details['market_cap'].capitalize()}")
            print(f"  Sustainability: {details['sustainability_score']*10}/10")
            print("Remember: This is not financial advice! Do your own research.")

        elif any(word in user_query for word in ['list', 'show', 'all', 'coins']):
            print("CryptoBuddy: Here are all the cryptocurrencies I track:")
            for coin in crypto_db:
                print(f"  {coin}:")
                print(f"    Trend: {crypto_db[coin]['price_trend'].capitalize()}")
                print(f"    Market Cap: {crypto_db[coin]['market_cap'].capitalize()}")
                print(f"    Sustainability: {crypto_db[coin]['sustainability_score']*10}/10")

        else:
            print("CryptoBuddy: I'm not sure I understand. Try asking about:")
            print("- Trending cryptocurrencies")
            print("- Sustainable/eco-friendly coins")
            print("- Long-term investment options")
            print("- High market cap coins")
            print("Or type 'list' to see all tracked cryptos")

# Start the chatbot
if __name__ == "__main__":
    crypto_advisor()
```
